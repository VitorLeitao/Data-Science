```python
import pandas as pd
import plotly.express as px
```

```python
df1 = pd.read_csv('/content/DataSets/NetflixOriginals.csv', encoding='ISO-8859-1') 
df2 = pd.read_csv('/content/DataSets/netflix_titles.csv')
```

```python
df1.head(5)
```

```python
df2.head(5)
```

vamos adicionar informações do d2 ao df1, sendo elas: 
- release_year
- country
- director	

```python
df = pd.merge(df1, df2[['title', 'release_year', 'country', 'director']], left_on='Title', right_on='title', how='left')
```

```python
df.sample(10)
```

Fazendo uma relação entre generos e notas

```python
dados = df.groupby('Genre')['IMDB Score'].mean().sort_values().reset_index()
```

```python
dados
```

```python
# grafico em barra para ter uma noção inicial das notas por genero
fig = px.bar(dados, x= 'Genre', y= 'IMDB Score')
fig.show()
```

```python
# Agora vamos fazer um box plot para termos uma ideia de quais generos não estão conseguindo performar de acordo com a media
fig = px.box(dados, y = 'IMDB Score')
fig.show()
```

```python
# Percebemos que temos 3 notas outliers, vamos ver quais são elas
dados[dados['IMDB Score'] == 8.2]
```

```python
dados[dados['IMDB Score'] == 3.9]
```

```python
dados[dados['IMDB Score'] == 3.7]
```

Feita essa primeira analise percebemos que os generos de filme que sao anormalmente mal avaliados são: 

- Heist film/Thriller
- Musical/Western/Fantasy	

em contrapartida, vemos que os generos de filmes que são anormalmente bem avaliados são: 
- Animation/Christmas/Comedy/Adventure	

Agora faremos a mesma anlise, mas para os paises, para saber em quais paises vale mais a pena investir e em quais não se vale a pena 

```python
dados = df.groupby('country')['IMDB Score'].mean().reset_index()
dados['count'] = df.groupby('country')['IMDB Score'].count().values
dados = dados.sort_values('IMDB Score', ascending=True)
```

```python
dados = dados[dados['count'] >= 3]
dados
```

```python
fig = px.bar(dados, x= 'country', y= 'IMDB Score')
fig.show()
```

```python
# veremos quais estao anormalmente fora da media
fig = px.box(dados, y = 'IMDB Score')
fig.show()
```

Nesse caso nao temos outliers pra cima, mas ainda podemos fazer uma analise com base nos que estão acima da media

```python

dados = dados[dados['IMDB Score'] >= 5.85]
dados
```

```python
fig = px.box(dados, y = 'IMDB Score')
fig.show() # entre os que estão acima da media, percebemos que temos um outliers, portanto, vamos analisa-lo
```

```python
dados[dados['IMDB Score'] == 7]
```

Percebemos que filmes do Reino Unido tem uma nota anormalmente acima da media entre os paises que ja estão acima da media

para fazer uma analise se vale a pena ou nao investir em filmes do reino unido, vamos ver se as suas notas:

- constantemente boas?
- decairam ou aumentaram com o passar do tempo?

```python
# Vamos ver a lista de notas dos filmes do reino unido
dados = df[df['country'] == 'United Kingdom']['IMDB Score'].reset_index().sort_values('IMDB Score', ascending=True)
dados
```

```python
fig = px.line(dados, x = 'index', y = 'IMDB Score')
fig.show()
```

```python
fig = px.box(dados, y = 'IMDB Score')
fig.show()
```

Por meio desses dois graficos, percebemos que nao ha uma difereça tao discrepante entre as notas dos filmes produzidos no reino unido

```python
# Analisando filmes do reino unido ao longo dos anos
dados = df[df['country'] == 'United Kingdom'].groupby('release_year')['IMDB Score'].mean().reset_index().sort_values('release_year', ascending=True)
dados
```

```python
fig = px.line(dados, x = 'release_year', y = 'IMDB Score')
fig.show()
```
