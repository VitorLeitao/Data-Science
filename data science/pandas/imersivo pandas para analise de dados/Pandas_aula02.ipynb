# Como fazer essa concatenação de forma mais facil?

```python
# Vamos iterar sobre esses arquivos da pasta(gera uma lista com esses arquivos)
arquivos = sorted(glob('/content/drive/MyDrive/IMERSIVO PANDAS/Planilhas/*.xlsx')) # Vai ler todos os arquivos .xlsx da planilha
```

```python
arquivos
```

```python
# Criando o df final de forma mais otimizada
df = pd.concat((pd.read_excel(cont) for cont in arquivos)) # cont vai ser cada um dos elementos da lista arquvios, concatenando todos eles
# o que ta dentro do concat vai retornar uma lista, uma vez que arquivos é uma lista
```

```python
df.sample(5)
```

```python
# Criando uma nova coluna 
# Precisamos de uma coluna para receita, uma vez que venda é o valor do produto e qtde a quantidade de produtos vendidos
df['Receita'] = df['Vendas'] * df['Qtde']
```

```python
df.sample(5)
```

```python
# Queremos outra coluna com o ano da venda(usando date_time)
df['Ano_venda'] = df['Data'].dt.year # Vai extrair o ano da data
```

```python
df.sample(5)
```

```python
df['Dia_venda'] = df['Data'].dt.day
df['Mes_venda'] = df['Data'].dt.month 
```

```python
df.sample(5)
```

# Como exportar o df que alteramos


```python
# Exportando dados para excel
df.to_excel("Planilha_att.xlsx", index=False) # Nao precisamso do index no excel
```
