```python
# Importando as bibliotecas
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from category_encoders import OneHotEncoder
```

```python
df = pd.read_csv('/content/drive/MyDrive/machine learning/dataset/titanic.csv')
```

```python
df.drop(["Name","Ticket", "Cabin", "Embarked", "PassengerId"], axis=1, inplace=True)
```

```python
df.sample(5)
```

```python
# dividindo em conjunto de treino e teste
X_train, X_test, y_train, y_test = train_test_split(df.drop(["Survived"], axis=1),
                                                   df["Survived"],
                                                   test_size=0.2,
                                                   random_state=42)
```

```python
# criando o modelo usando pipeline
model = Pipeline(steps=[
    ('one-hot encoder', OneHotEncoder()), # Botar as variaveis como numericas(equivalente ao map)
    ('Imputer', SimpleImputer(strategy='mean')), # Subistituindo os valores ausentes pela media
    ('Tree', DecisionTreeClassifier(random_state=42)) # vai instanciar o metodo pra classificação(ultimo passo sempre será esse!!)
])
```

```python
# Treinando nosso modelo

model.fit(X_train, y_train)
```

```python
# Avaliação do modelo
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)
```
