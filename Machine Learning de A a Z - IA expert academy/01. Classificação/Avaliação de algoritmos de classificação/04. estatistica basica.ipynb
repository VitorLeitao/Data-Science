# Variancia

é a media da diferença de cada valor pela media elevado ao quadrado
```
Valores = [150, 151, 152, 153, 154, 155, 155, 155]

media = 153

desvios = [3,2,1,1,0, 1, 2, 2, 2]

variancia vai ser (for i in n: desvios[i] ** 2) / n

variancia = 28 / 9
variancia = 3,11
```
- A variancia deixa mais evidente o quão longe os valores estão do esperado, uma vez que eleva tudo ao quadrado
- quanto menor o valor da variancia, mais preciso é o algoritmo

# Desvio padrão

- é um complemento da variancia
```
desvio_padrao = sqrl(variancia)
```
- mais interpretável e completo
- indica basuicamente o erro, exemplo
- DP = X, os valores podem se afastar da media X quantidada para direia ou para a esquerda

# Coeficiente de variação

- indica a porcentagem de variação da base

```
CV = (desvio padrao / Media) * 100
cv = (1,76 / 153) * 100
CV = 1,15%
```

## Teste de hipotese

- teste de hipotese é a validação ou rejeitação de uma frase para uma amostra de uma população
- pegamos a media e determinamos intervalos de confiança, ex: media de idade é 20, deixamos um intervalo entre 15 e 25

```
hipotese : ideia a ser testada -> ex: queremos melhorar um algoritmo X

hipotese nula(h0) : informação que ja existia -> por exemplo, o resultado do algoritmo é 20, presumimos qque ela é verdadeira ate que se prove o contrario

hipotese alternativa(h1): o que está tentando provar, tudo o que é diferente de h0 -> ex: o valor certo é 10

Alpha(confiabilidade): Probabilidade maxima de rejeitar a hipótese nula, quanto menor mais seguro é o resultado (nível de significância) - em geral 0.01 ou 0,05

p-value: p >= alph: nao rejeita h0(nao temos evidencias)

erro tipo 1: rejeitar h0 quando nao deveria

erro tipo 2: nao rejeitar a h0 quando deveria ter rejeitado
```

**serve para verificar se há uma diferença significativa entre grupos**

- comparação entre 3 ou mais grupos(amostras independentes) -> venda de produtos em 3 dias diferentes
- uma variavel quantitativa e uma ou mais quantitativas
- dados precisam estar na distribuição normal
- compara a variação do grupo X com a variação do grupo Y e Z


```
ho: não há diferença estatistica significativa entre os conjuntos
h1: há uma diferença estatistica significativa
```

```
IF(F estiver entre 0 e F critico):
    não existe diferença estatistica

ELSE:
    existe diferença estatistica
```
