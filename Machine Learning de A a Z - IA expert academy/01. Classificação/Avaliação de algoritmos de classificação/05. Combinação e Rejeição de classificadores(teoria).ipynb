# Combinação de classificadores

- Podemos usar mais de um classificador, buscando o melhor conjunto de classificadores para combinar
- classificação mais precisa
- diminui o risco de selecionar um classificador individual inadequado


```
unanimity -> todos os algoritimos tem que dar a mesma resposta para classificar

simple majority -> 50% + 1 classificadores consideramos a resposta final

plutality -> o que tiver mais voto é a resposta correta
```



# Rejeição de classificadores

- Rejeita o resultado se não tiver certeza suficiente
- Se tivermos uma porcentagem que nao seja confiavel para uma classe



```
1. classificador da uma porcentagem de certeza para cada classe
2. Analisa-se se a hipotese irá ser rejeitado ou não
```
