# SVM 
teórico

- Em geral supera outros algoritmos de ML
- principalmente em tarefas mais complexas como caracteres, voz, imagens


**funcionamento base**
- Aprende hiperplanos de separação com margem máxima, sendo essa margem a distancia entre a reta do grafico e dos elementos
- a reta divide os elementos em classes
- se tivermos uma margem maxima grande, as classes ficam mais distantes uma da outra, sendo mais facil de encontrar a classe de um novo elemento, generalizando melhor os elementos


**como separar hiperplanos**
- Podemos usar o convex hulls para gerar hiperplanos, ou seja, as classes, separadas por uma reta
- Também podemos usar abordagem matematica, usando a formula do hiperplano de separação, criando uma reta parcial, e apos isso a reta final

**adiministração de erros**
- o SVM pode variar sua taxa de erro a partir de uma variavel C
- caso C seja alto, tenta ao maximo criar uma reta que separe perfeitamente os dados, sem nenhum erro
- caso C seja baixo, o algoritmo permitirá mais erros
- assim como tudo em ML, temos que ir testando e ver qual fica melhor para nossa base de dados

**linear X não linear**
- Lineares são os problemas no qual podemos traçar uma reta e separar as classes, enquanto nao lineares sao aqueles onde isso não é possivel

**SVM não lineares**

- kernel trick
- reorganizaremos o grafico usando kernel tricks de modo a fazer com que reorganizemos os dados para transformar o problema em linear
- as formas de fazer isso são muito variadas, uma forma é transformar o plano de R1 para R2, adicionando uma dimensão 
- Ele consegue criar novos atributos de modo a transformar o sistema em linar


**vantagens**
- pouco influenciado por ruidos
- classificação e regressão
- aprende conceitos nao presentes nos dados originais, por exemplo criar um atributo altitude baseado na longitude e latitude
- mais fácil de usar do que redes neurais

**desvantagens**
- precisa testar varias combinações
- Lento
- Black Box - não interpretável

## Base credit - 0.988

```python
from sklearn.svm import SVC
```

```python
import pickle
with open('/content/drive/MyDrive/Machine Learning e Data Science com Python de A a Z/01. Classificação /credit.pkl', 'rb') as f:
  X_credit_train, X_credit_test, y_credit_train, y_credit_test = pickle.load(f)
```

```python
X_credit_train.shape, y_credit_train.shape,   X_credit_test.shape, y_credit_test.shape
```

```python
svm_credit = SVC(kernel='rbf', random_state=1, C = 2.0)
svm_credit.fit(X_credit_train, y_credit_train)
```

```python
previsoes = svm_credit.predict(X_credit_test)
```

```python
previsoes
```

```python
y_credit_test
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(y_credit_test, previsoes))
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(svm_credit)
cm.fit(X_credit_train, y_credit_train)
cm.score(X_credit_test, y_credit_test)
```

```python
print(classification_report(y_credit_test, previsoes))
```
