# Aprendizagem por regras
teorico

- vamos ter regras, semelhantes a "ifs", alem disso, teremos uma regra default, que é equivalente ao "else", se nao encaixamos em nenhuma regra, jogaremos no default


**Algoritmo OneR**
- testa coisas simples primeiro
- Um atributo faz todo trabalho
- vai fazer uma grade tabela, onde para cada atributo ele determina  varias regras, calcula a taxa de erro de cada atributo, e pega o atributo com menor taxa de erro e deixa ele como principal, ou seja, usaremos apenas as regras que usam esse atributo para determinar as previsões

**Algoritmo PRISM**
- Faremos uma regra para cada valor unico do atributo alvo
- Se ? entao X, Se ? entao Y, ...
- para calcular o ?, faremos por abrangencia, ou seja, para cada valor unico do atributo alvo, pegaremos a valor dos outros atributos que aparecem em mais linhas e colocaremos eles nas regras
- para risco = alto, temos uma abrangencia maior de historico de cregito = ruim, divida = alta, renda <= 15, ou seja, para as linhas onde risco = alto, os atributos apresentam em sua maioria as caracteristicas acima, logo
- se divida = alta, renda <= 15 , historico = ruim, entao risco = alto

dividindo a base de dados em treino e teste

```python
base_dividida = Orange.evaluation.testing.sample(base_credit, n = 0.25) # Treino e teste
base_train = base_dividida[1]
base_test = base_dividida[0] 
len(base_test), len(base_train)
```

```python
cn2 = Orange.classification.rules.CN2Learner()
regras_credit = cn2(base_train)
```

```python
for c in regras_credit.rule_list: # Vai retornar a lista de regras
  print(c)
```

```python
previsoes = Orange.evaluation.testing.TestOnTestData(base_train, base_test, [lambda testdata: regras_credit])
previsoes
```

```python
Orange.evaluation.CA(previsoes)
```

```python
Orange.evaluation.F1(previsoes)
```

#Classificador base - Majority leaner

- algoritmo que vai fazer a classificação dos registros baseado na maioria dos dados na base de dados, usado para verificar a eficácia dos algoritmops de ML
- Indica um percentual minimo para os algoritmos de ML

## Base credit data

```python
base_credit = Orange.data.Table('/content/drive/MyDrive/Machine Learning e Data Science com Python de A a Z/Bases de dados/credit_data_regras.csv')
base_credit
```

```python
base_credit.domain
```

```python
majority = Orange.classification.MajorityLearner()
```

```python
previsoes = Orange.evaluation.testing.TestOnTestData(base_credit, base_credit, [majority]) # "treinando" e testando na mesma base 
```

```python
Orange.evaluation.CA(previsoes)
```
